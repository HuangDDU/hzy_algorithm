查看结果

```python
def show_map(map):
    N = map.N

    for i in range(N):
        for j in range(N):
            node = map.node_matrix[i][j]
            print(node.distance, end=" ")
        print()

    print("==================================")
    for i in range(N):
        for j in range(N):
            node = map.node_matrix[i][j]
            if node.best_direction == "UP":
                print("↑", end=" ")
            elif node.best_direction == "RIGHT":
                print("→", end=" ")            
            elif node.best_direction == "DOWN":
                print("↓", end=" ")            
            elif node.best_direction == "LEFT":
                print("←", end=" ")
            else:
                print("U", end=" ")
        print()
```

```python
map_filename = "main18_format_transform_weight.txt"
```

旧格式转换思路，每个transmitter需要的格式取出现次数最多的

```python
map, provider, consumer_vector, transmitter_vector = main_old(conf_file=map_filename)
visualization(map, provider, consumer_vector, transmitter_vector)
show_map(map)
```

新格式转换思路，每个transmitter尝试输入格式计算代价，选择代价最小的

```python
map, provider, consumer_vector, transmitter_vector = main(conf_file=map_filename)
visualization(map, provider, consumer_vector, transmitter_vector)
show_map(map)
```
