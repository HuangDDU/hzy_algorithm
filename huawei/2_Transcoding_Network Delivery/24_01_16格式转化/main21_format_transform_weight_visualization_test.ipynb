查看结果

```python
def show_map(map):
    N = map.N

    for i in range(N):
        for j in range(N):
            node = map.node_matrix[i][j]
            print(node.distance, end=" ")
        print()

    print("==================================")
    for i in range(N):
        for j in range(N):
            node = map.node_matrix[i][j]
            if node.best_direction == "UP":
                print("↑", end=" ")
            elif node.best_direction == "RIGHT":
                print("→", end=" ")            
            elif node.best_direction == "DOWN":
                print("↓", end=" ")            
            elif node.best_direction == "LEFT":
                print("←", end=" ")
            else:
                print("U", end=" ")
        print()
```

```python
map_filename = "main21_summary.txt"
```

暴力格式转换，到最后的Consumer才转换

```python
map, provider, consumer_vector, transmitter_vector = main_format1(conf_file=map_filename)
visualization(map, provider, consumer_vector, transmitter_vector)
show_map(map)
```
